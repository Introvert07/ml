```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.txt',sep = ';',header =None,names = ['text','emotion'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['emotion'].unique()
```

```python
unique_emotions = df['emotion'].unique()
emotion_numbers = {}
i=0
for emo in unique_emotions:
    emotion_numbers[emo] = i
    i+=1
df['emotion'] = df['emotion'].map(emotion_numbers)
```

```python
df.head()
```

```python
df['text'] = df['text'].apply(lambda x :x.lower())
```

```python
import string

def remove_punc(txt):
    txt = txt.translate(str.maketrans('','',string.punctuation))
    return txt
```

```python
df['text'] = df['text'].apply(remove_punc)
```

```python
def remove_numbers(txt):
  new = ""
  for i in txt:
    if not i.isdigit():
      new = new+i
  return new
df['text'] = df['text'].apply(remove_numbers)
```

```python
def remove_emojis(txt):
  new = ""
  for i in txt:
    if i.isascii():
          new += i
  return new
df['text'] = df['text'].apply(remove_emojis)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python
len(stop_words)
```

```python
df.loc[1]['text']
```

```python
def remove(txt):
  words = txt.split()
  cleaned = []
  for i in words:
    if not i in stop_words:
      cleaned.append(i)

      return ''.join(cleaned)
```

```python
df['text'] = df['text'].apply(remove)
```

```python
df.loc[1]['text']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
documents = [
    "I love Pizza",
    "Pizza is the best",
    "I love Pizza",
    "Pasta is great"
]
vectorizer = CountVectorizer(ngram_range=(2,2))
X = vectorizer.fit_transform(documents)


print("vocab:", vectorizer.get_feature_names_out())
print("\nBoW Matrix:\n", X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents = [
    "I love Pizza",
    "Pizza is the best",
    "I love Pizza",
    "Pasta is great"
]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)


print("vocab:", vectorizer.get_feature_names_out())
print("\nBoW Matrix:\n", X.toarray())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)


nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)


pred_bow = nb_model.predict(X_test_bow)
print(accuracy_score(y_test, pred_bow))
```

```python
pred_bow
```

```python
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)


nb2_model = MultinomialNB()
nb2_model.fit(X_train_tfidf,y_train)
```

```python
y_pred = nb2_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
logictic_model = LogisticRegression(max_iter=1000)
logictic_model.fit(X_train_tfidf,y_train)
```

```python
log_pred = logictic_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test,log_pred))
```
